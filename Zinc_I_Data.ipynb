```python
from padelpy import padeldescriptor
from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# kfkr = padel_fingerprints_xml[7]
fingerprint_output_file = 'zinc_kkrf_output.csv'
klekotaRothFingerprintCount_xml = padel_fingerprints_xml[7]

# kkrc_fp = pd.read_csv('KlekotaRothCount.csv')
# X = kkrc_fp.drop('Name', axis = 1)

# selector = VarianceThreshold()
# X = selector.fit_transform(X)
# X = pd.DataFrame(X)


padeldescriptor(mol_dir='zinc_lead_mol_data.smi', 
                d_file=fingerprint_output_file, 
                descriptortypes= klekotaRothFingerprintCount_xml,
                detectaromaticity=True,
                standardizenitro=True,
                standardizetautomers=True,
                threads=2,
                removesalt=True,
                log=True,
                fingerprints=True
               )
```

```python
import pickle
# two models:  serotonin_model_confirm_kkr.pkl and serotonin_model.pkl

with open('serotonin_model_confirm_kkr.pkl', 'rb') as serotonin_model:
    model = pickle.load(serotonin_model)
    
y_predict = model.predict(X)
y_predict
```

```python
ydf= pd.DataFrame({'Ki': y_predict})
ydf.to_csv('result3/zinc_mol_predicted_ki.csv', index=False)
```
