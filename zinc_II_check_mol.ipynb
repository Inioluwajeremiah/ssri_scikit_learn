```python
# Generate Molecular FIngerprint using Padel Descriptors
import pandas as pd

zinc_smiles= pd.read_csv('zinc_lead_smiles.csv')

from padelpy import padeldescriptor
 
    
fingerprint_descriptor_name = 'Zinc_KlekotaRoth'
fingerprint_descriptor_xml = padel_fingerprints_xml[7]
fingerprint_output_file = ''.join([fingerprint_descriptor_name , '.csv'])
    
padeldescriptor(mol_dir='zinc_lead_mol_data.smi', 
                d_file=fingerprint_output_file, 
                descriptortypes= fingerprint_descriptor_xml,
                detectaromaticity=True,
                standardizenitro=True,
                standardizetautomers=True,
                threads=2,
                removesalt=True,
                log=True,
                fingerprints=True)


descriptors = pd.read_csv(fingerprint_output_file)
descriptors
```

```python
# check using fingerprint if the best 50 ligands are not in the training set used in building the model
import pandas as pd
get_zinc_smiles = pd.read_csv('zinc_lead_smiles.csv')
get_best_50_smiles = pd.read_csv('result3/combined_zinc_ki_best_50.csv')
# kkrc_fp
get_best_50_smiles
```

```python
unrepeated_smile = []
unrepeated_id = []
for index in range(0,50):
    if get_best_50_smiles['molecule_smile'][index] not in get_zinc_smiles['molecule_smile']:
        unrepeated_smile.append(get_best_50_smiles['molecule_smile'][index])
        unrepeated_id.append(get_best_50_smiles['Name'][index])
        
        # save each smile in a .smi format
        f = open(f"top_smile/{get_best_50_smiles['Name'][index]}.smi", "w")
        f.write(f"{get_best_50_smiles['molecule_smile'][index]}")
        f.close()

        

unrepeated_df = pd.DataFrame({'id': unrepeated_id, "smile":unrepeated_smile})
unrepeated_df.to_csv('result3/confirmed_unrepeated_zinc_smi.csv', index=False)
unrepeated_df
```
