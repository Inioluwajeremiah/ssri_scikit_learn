```python
#Loading the dataset
import pandas as pd
from chembl_webresource_client.new_client import new_client

serotonin_target = new_client.target
serotonin_targets = serotonin_target.search('serotonin')
sert_mm_df = pd.DataFrame.from_dict(serotonin_targets)

sert_mm_df.head()

# save seotonin macromelecule targets to csv
# sert_mm_df.to_csv("serotonin_targets.csv", index=False)
```

```python
# select molecule id
sert_mm_target = sert_mm_df.target_chembl_id[2]
sert_mm_target
```

```python
# we first searched for a specific serotonin target. Serotonin targets (they are the macromolecule) according to chembl, are grouped into different
# categories, so we selected a particular category of interest (CHEMBL228). This target is serotonin transporter (a protein mac
# romolecule that transports serotonin in human). This target had been worked on with different serotonin inhibitors
# e.g target_organism = human, target_type=single_protein, 

serotonin_activity = new_client.activity
sert_mm_by_Ki = serotonin_activity.filter(target_chembl_id = sert_mm_target).filter(standard_type="Ki")
sert_mm_by_ki_df= pd.DataFrame.from_dict(sert_mm_by_Ki)
# sert_mm_by_ki_df.to_csv("serotonin_bioactivity.csv", index=False)
print(sert_mm_by_ki_df.shape)
sert_mm_by_ki_df.head()
```

```python
# clean data set

clean_standard_value = sert_mm_by_ki_df[sert_mm_by_ki_df.standard_value.notna()]
clean_canonical_smiles = clean_standard_value[clean_standard_value.canonical_smiles.notna()]
clean_repeated_molecules = clean_canonical_smiles.drop_duplicates(subset=['molecule_chembl_id'], keep= 'first')
# clean_canonical_smiles.to_csv("serotonin_target_clean_dataset.csv", index=False)
print('clean_standard_value', clean_standard_value.shape)
print('clean_canonical_smiles', clean_canonical_smiles.shape)
print('clean_repeated_molecules', clean_repeated_molecules.shape)
clean_repeated_molecules.to_csv("clean_serotonin_tragets.csv", index=False)
```

```python
import pandas as pd

serotonin_target_clean_dataset = pd.read_csv('clean_serotonin_tragets.csv')
clean_serotonin_tragets = serotonin_target_clean_dataset[['molecule_chembl_id', 'canonical_smiles', 'standard_value']]

clean_serotonin_tragets.to_csv('serotonin_clean_needed_dataset.csv', index=False)
serotonin_target_clean_dataset.head()
clean_serotonin_tragets
```
